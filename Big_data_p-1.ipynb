```python
with open('data.txt') as data_file:
    data = data_file.read()
    print(data)
```

```python
data = data.replace(':','')
l = data.split()
hadoop_position, java_position = [], []

for index, i in enumerate(l):
  if i.lower() == 'hadoop' :
    hadoop_position.append(index)
  if i.lower() == 'java' : 
    java_position.append(index) 
print(f'Hadoop Positions: {hadoop_position}, frequency: {len(hadoop_position)}')
print(f'Java Positions: {java_position}, frequency: {len(java_position)}')
```
